```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# sample data
data = {
    'Hero' :['Iron Man', 'tor ', 'catin meica', 'blkca pater'],
    'Strenght' : [85, 95, 86, 90],
    'Speed' : [97, 89, 98, 79 ],
    'Intellegence' : [96, 80, 96, 57]
}

df = pd.DataFrame(data)
print(df)
```

```python
# bar plot - strenght of each super hero
plt.figure(figsize=(8,5))
sns.barplot(x = 'Hero', y = 'Strenght', data =df)
plt.title('SuperHero Strenght')
plt.xlabel('Hero')
plt.ylabel('Strenght Level')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize= (8,5))
plt.plot(df['Hero'], df['Speed'], label = 'Speed', marker = 'o')
plt.plot(df['Hero'], df['Intellegence'], label = 'Intellegence', marker = 's')
plt.title("Speed vs Intellegence")
plt.xlabel('Hero')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
corr = df[['Strenght', 'Speed', 'Intellegence']].corr()
plt.figure(figsize= (6,4))
sns.heatmap(corr)
plt.title('Stat Correlation')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.animation as animation
fig, ax = plt.subplots()
ax.set_xlim(0,2*np.pi)
ax.set_ylim(-1.5, 1.5)
#initialize an empty line on the plot
line, = ax.plot([],[], lw=2, color='dodgerblue')
def init():
  line.set_data([],[])
  return line,
def animate(i):
  x= np.linspace(0,2*np.pi, 100)
  y = np.sin(x + i / 10.0)
  line.set_data(x,y)
  return line,
ani = animation.FuncAnimation(fig,animate,init_func=init,frames=100,interval=50,blit=True)
from IPython.display import HTML
HTML(ani.to_jshtml())
```

```python
from ColabTurtle.Turtle import*
initializeTurtle()
speed(5)
pensize(2)
color("purple")

#functino to create petal
def draw_petal():
  for _ in range(18): # draw first curve
      forward(5)
      left(10)
  left(90)
  for _ in range(10):
    forward(10)
    left(10)
  left(90)
for _ in range(6):
  draw_petal()
  left(60)


```
